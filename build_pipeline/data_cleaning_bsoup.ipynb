```python
import requests
import IPython
```

```python
response = requests.get(r"https://www.udacity.com/courses/all")
with open("data/html.txt", 'w', encoding = 'utf-8') as f:
    f.write(response.text)
```

```python
import re

pattern = re.compile(r'<.*?>')

print(pattern.sub('', response.text))
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(response.text, 'html.parser')
print(soup.get_text())
```

```python
exp = re.compile(r'filter-mobile_subsectionOption__*')
courses = soup.find_all("button",{"class": exp})
```

```python
course_list = [course.get_text().strip() for course in courses]
```
